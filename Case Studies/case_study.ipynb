```python
# Import Libraries
import pandas as pd

# Load the Excel file
file_path = 'Log intern - Analyst case study.xlsx'

# Read the two sheets
brands = pd.read_excel(file_path, sheet_name='Data brands')
orders = pd.read_excel(file_path, sheet_name='Data orders')
```

```python
# Define functions
def merge_datasets(brands,orders):
    # Join the two dataframes
    final_datastet = pd.merge(brands,orders,how='inner',on='uuid_brand')
    return final_datastet

def get_brand_orders(dataset, state):
    if state:
        # Filter for synced brands
        filtered_dataset = dataset[dataset['fg_ff_sync*'] == state]

        brand_orders_dict = {}
        for brand in filtered_dataset['uuid_brand'].unique():
            brand_orders = filtered_dataset[filtered_dataset['uuid_brand'] == brand]['Order reference'].tolist()
            brand_orders_dict[brand] = brand_orders
        return brand_orders_dict
    else:
        raise ValueError("Please enter a sync state (true or false) to view dataset.")

```

```python
# Call functions to output dictionary of synced brands and their orders
get_brand_orders(merge_datasets(brands,orders), True)
```
