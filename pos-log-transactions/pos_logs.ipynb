```python
import pandas as pd
```

```python
#Load all pos-operator-logs into dataframes.
log1 = pd.read_csv("./POS_operator_logs_20171207-20171219.csv")
log2 = pd.read_csv("./POS_operator_logs_20190213-20190226.csv")
log3 = pd.read_csv("./POS_operator_logs_20190328-20190410.csv")
```

```python
#Merge all 3 log dataframes
pos_logs = pd.concat([log1,log2,log3])
```

```python
pos_logs.head()
```

```python
#rename columns
pos_logs.rename(columns= {"WorkstationGroupID":"workstationgroupid","WorkstationID":"workstationid","TranID":"tranid","BeginDateTime":"start","OperatorID":"opid","Items":"opstat"}, inplace=True)
```
