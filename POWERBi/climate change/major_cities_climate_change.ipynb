```python
majorcity_temps =  pd.read_csv(r"C:\\Users\\brian\\Desktop\\portfolio\\climate_change\\original\\GlobalLandTemperatures\\GlobalLandTemperaturesByMajorCity.csv")
majorcity_temps.head()
```

```python
len(majorcity_temps['Country'].unique())
```

```python
highest_temp = majorcity_temps['AverageTemperature'].max()
highest_temp
```

```python
lowest_temp = majorcity_temps['AverageTemperature'].min()
lowest_temp
```

```python
highest_majorcity_temp_date = majorcity_temps[majorcity_temps['AverageTemperature'] == highest_temp]
highest_majorcity_temp_date
```

```python
lowest_majorcity_temp_date = majorcity_temps[majorcity_temps['AverageTemperature'] == lowest_temp]
lowest_majorcity_temp_date
```

```python
top_20_hottest_cities = majorcity_temps.sort_values('AverageTemperature',ascending=False)
top_20_hottest_cities.head(20)
```

```python
top_20_coldest_cities = majorcity_temps.sort_values('AverageTemperature',ascending=True)
top_20_coldest_cities.head(20)
```
