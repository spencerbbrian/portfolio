```python
#Import all libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
cylicist = pd.read_csv('cylicist.csv')
cylicist
```

```python
# Check for NA values per attribute
cylicist.isna().sum()
```

```python
# Check for null values per attribute
cylicist.isnull().sum()
```

```python
#Drop all null/Na values
cylicist = cylicist.dropna(how="all")
cylicist.dropna(how="any", inplace=True)
cylicist
```

```python
#Cross check drop of null values
cylicist.isnull().sum()
```

```python
# Get all unique values per attribute
print(cylicist['rideable_type'].unique())
print(cylicist['start_station_name'].unique())
print(cylicist['end_station_name'].unique())
print(cylicist['member_casual'].unique())
```
