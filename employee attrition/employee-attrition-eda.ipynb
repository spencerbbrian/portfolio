## Employee Attrition Exploratory Data Analysis

```python
#import libraries
import pandas as pd
import seaborn as sns
sns.set_style('dark')
sns.set_palette('RdBu')
import matplotlib.pyplot as plt
```

```python
#Load the data
attrition = pd.read_csv('./employee-attrition-main.csv')
attrition.head()
```

```python
#Drop Unnecessary Columns
attrition.drop(columns=["EmployeeCount","StandardHours"]).head()
```

```python
#Check for null values
attrition.isnull().sum()
```

```python
#Get information of columns
attrition.info()
```

```python
#Get description of dataset columns
attrition.describe()
```

```python
#Get shape of dataset
attrition.shape
```

```python
sns.heatmap(pd.crosstab(attrition['YearsInCurrentRole'],attrition['YearsSinceLastPromotion'],values=attrition['MonthlyIncome'],aggfunc="mean"))
```

```python
#Categorical Visualization of Attrition
category_order = ["Yes","No"]
sns.catplot(x="Attrition",data = attrition,kind="count",order=category_order)
plt.show()
```

```python
#Relational Plot of Attrition vs Monthly Income
sns.relplot(x="YearsAtCompany",y="MonthlyIncome",data=attrition,kind="scatter",hue="YearsSinceLastPromotion")
plt.show()
```

```python
sns.catplot(x="Attrition",y="MonthlyIncome",data=attrition,kind="box")
plt.show()
```

```python
#Create a dictionary for the Education Field Table
index = [1,2,3,4,5,6]
education_field_id = ["EF10","EF11","EF12","EF13","EF14","EF15"]
education_field_name = ["Human Resource","Technical Degree","Life Sciences","Medical","Marketing","Other"]
dict = {education_field_id[i] : education_field_name[i] for i in range(len(education_field_id))}
education_field = pd.DataFrame([education_field_id,education_field_name]).transpose()
education_field.rename(columns={0:"education_field_id",1:"education_field_name"},inplace=True)
education_field
```

```python
#Split main dataset into tables
education = attrition[["EmployeeNumber","Education","EducationField"]]
benefits = attrition[["EmployeeNumber","DailyRate","HourlyRate","MonthlyRate","MonthlyIncome","PercentSalaryHike"]]
family = attrition[["EmployeeNumber","MaritalStatus","DistanceFromHome"]]
attrition_main = attrition[["EmployeeNumber","Age","Attrition","Department","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager","TotalWorkingYears","TrainingTimesLastYear","Gender","OverTime","PerformanceRating","JobInvolvement","JobLevel","JobRole"]]
```

```python
#Get new csvs for tables
education.to_csv('education.csv')
benefits.to_csv('benefits.csv')
family.to_csv('family.csv')
attrition_main.to_csv("attrition.csv")
education_field.to_csv("education_field.csv")
```
