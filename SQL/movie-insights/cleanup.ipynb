```python
# Install Libraries
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
# Find the number of missing values in each column
movies.isnull().sum()
```

```python
# Drop the rows where at least one element is missing
movies_cleaned = movies.dropna()
movies_cleaned.isnull().sum()
```

```python
# Crosscheck there is enough data to work with
movies_cleaned.count()
```

```python
# Split the released column into release date and release country
movies_cleaned[['release_date','release_country']] = movies_cleaned['released'].str.split('(',expand=True)
movies_cleaned['release_country'] = movies_cleaned['release_country'].str.replace(')','')
movies_cleaned
```

```python
# Delete redundant columns
movies_cleaned.drop(['released'],inplace=True,axis=1)
movies_cleaned.head()
```

```python
# Convert the new cleaned dataset to a new csv file
movies_cleaned.to_csv('movies_cleaned.csv', index=False)
```
