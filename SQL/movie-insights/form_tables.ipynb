```python
import pandas as pd
```

```python
movie_insights = pd.read_csv('movies_cleaned.csv')
movie_insights.head()
```

```python
# Create Directors table
directors = movie_insights[['director']].drop_duplicates()
# Verify the directors table
directors.index = range(1, len(directors) + 1)
directors.to_csv('directors.csv', index=True, header=False, index_label='Index')
```

```python
# Create Genres table
genres = movie_insights[['genre']].drop_duplicates()
# Verify the genres table
genres.index = range(1, len(genres) + 1)
genres.to_csv('genres.csv', index=True, header=False, index_label='Index')
```

```python
# Create Ratings table
ratings = movie_insights[['rating']].drop_duplicates()
# Verify the ratings table
ratings.index = range(1, len(ratings) + 1)
ratings.to_csv('ratings.csv', index=True, header=False, index_label='Index')
```

```python
# Create Companies table
companies = movie_insights[['company']].drop_duplicates()
# Verify the companies table
companies['company_id'] = range(1, len(companies) + 1)

company_map = companies.set_index('company')['company_id'].to_dict()
companies.to_csv('companies.csv', index=False, header=True)
```

```python
# Create Countries table
countries = movie_insights[['country']].drop_duplicates()
# Verify the countries table
countries['country_id'] = range(1, len(countries) + 1)

country_map = countries.set_index('country')['country_id'].to_dict()
countries.to_csv('countries.csv', index=False, header=True)
```

```python
# Create Stars table
stars = movie_insights[['star']].drop_duplicates()
# Verify the stars table
stars.index = range(1, len(stars) + 1)
stars.to_csv('stars.csv', index=True, header=False, index_label='Index')
```

```python
# Create writers table
writers = movie_insights[['writer']].drop_duplicates()
# Verify the writers table
writers.index = range(1, len(writers) + 1)
writers.to_csv('writers.csv', index=True, header=False, index_label='Index')
```

```python
# Create Movies table
movies = movie_insights[['name','rating','year','score','votes','country','company','budget','gross','runtime','release_date','release_country']]
movies.head()
```

```python
# Map dictionaries to movies table
movies['company_id'] = movies['company'].map(company_map)
movies['country_id'] = movies['country'].map(country_map)
movies = movies.drop(columns=['company', 'country'])
movies.to_csv('movies_final.csv', index=False, header=True)
```
