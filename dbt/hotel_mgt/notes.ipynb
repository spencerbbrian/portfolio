1. The Best Data Strategy: "Synthetic-Real" Hybrid
Finding a single "perfect" open-source dataset for a multi-hotel franchise is tough because real hotel data is highly proprietary. I recommend a hybrid approach:

A. Use a High-Quality Base Dataset
Start with the Hotel Booking Demand Dataset (available on Kaggle).

Why: It contains ~119,000 observations for a City Hotel and a Resort Hotel.

Attributes: It includes ADR (Average Daily Rate), lead time, arrival dates, stay duration, adults/children, meal plans, and booking cancellations.

The "Franchise" Twist: You can use Python to "clone" this data and slightly modify the values (using a random multiplier for prices/occupancy) to create 5–10 different "properties" under your franchise.

B. Enrich with Python (Faker Library)
To make it look like a modern franchise, use a Python script to generate supplementary tables:

Property Metadata: Hotel names, locations (global cities), room counts, and tier (Economy vs. Luxury).

Loyalty Data: Generate a loyalty_members table with join dates and tiers (Gold, Platinum).

Operating Costs: Generate a table of daily labor, utility, and maintenance costs per hotel so you can calculate Net Profit instead of just Revenue.

2. Recommended Data Schema
To impress as an Analytics Engineer, your Snowflake internal stages should lead to a structured Dimensional Model (Star Schema).

Table Type,Name,Key Columns
Fact,fact_bookings,"booking_id, hotel_id, guest_id, check_in_date, revenue, stays_nights"
Dimension,dim_hotels,"hotel_id, hotel_name, city, country, star_rating, total_rooms"
Dimension,dim_guests,"guest_id, country_of_origin, loyalty_tier, first_booking_date"
Fact (Daily),fact_hotel_daily_stats,"hotel_id, date, rooms_occupied, cleaning_costs, electricity_costs"

3. High-Value KPIs to Model in dbt
Hiring managers love seeing industry-standard metrics. Ensure your dbt models calculate these:

RevPAR (Revenue Per Available Room): Total Room Revenue / Total Rooms Available. This is the "North Star" metric for hotels.

ADR (Average Daily Rate): Total Room Revenue / Number of Rooms Sold.

Occupancy Rate: (Rooms Sold / Rooms Available) * 100.

Lead Time Analysis: The average days between booking and arrival (crucial for marketing).

Cancellation Rate: To help the franchise understand revenue leakage.
